```
import pandas as pd 
import numpy as np
from numpy import nan
from numpy import array
from nltk.tokenize import TreebankWordTokenizer
from keras.layers import Input
from keras.models import Model
from keras.layers import Dense
data = pd.read_csv('/content/drive/My Drive/annotatedData.csv') 

```

```
data=data.to_numpy()
```

```
x,y=data.shape
```

```
x=10000
```

```
arr = [[]]
arrclass=[[]]
```

```
j=0
i=0
while(i<x):
    arr[j].append(data[i][1])
    arrclass[j].append(data[i][2])
    if(i+1!=x and data[i+1][0]=="na"):
      j=j+1
      i=i+1
      arr.append([]) 
      arrclass.append([])
    i=i+1  
```

```
arr
```

```
uniquewords=[]
uniqueclasses=[]
for i in range(len(arr)):
  for j in range(len(arr[i])):
    if(arr[i][j] not in uniquewords):
      uniquewords.append(arr[i][j])
for i in range(len(arrclass)):
  for j in range(len(arrclass[i])):
    if(arrclass[i][j] not in uniqueclasses):
      uniqueclasses.append(arrclass[i][j])      
```

```
uniquewords
```

```
cc = pd.DataFrame(np.zeros((len(uniqueclasses), len(uniqueclasses))))
cc.index=uniqueclasses
cc.columns=uniqueclasses
for i in range(cc.shape[0]):
  for j in range(cc.shape[1]):
    if(uniqueclasses[i]==uniqueclasses[j]):
      cc.loc[uniqueclasses[i],uniqueclasses[j]]=1
```

```
c = pd.DataFrame(np.zeros((len(uniquewords), len(uniquewords))))
c.index=uniquewords
c.columns=uniquewords
for i in range(c.shape[0]):
  for j in range(c.shape[1]):
    if(uniquewords[i]==uniquewords[j]):
      c.loc[uniquewords[i],uniquewords[j]]=1
window=5
#k=(window-1)/2
tes=[0]*len(uniquewords)
trainX=[]
trainY=[]
for v_s in range(len(arr)):
  for s_v in range(len(arr[v_s])):
     tes=[0]*len(uniquewords)
     for s in range(window):
         if(0<=s_v-2+s  and len(arr[v_s])>s_v-2+s ):
           wor=arr[v_s][s_v-2+s]
           ind=uniquewords.index(wor)
           tes=tes+c[uniquewords[ind]].to_numpy()
     trainX.append(tes) 
     ix=uniqueclasses.index(arrclass[v_s][s_v])
     trainY.append(cc[uniqueclasses[ix]].to_numpy())     
```

```
trainX
```

```
trainY
```

```

def define_model1(uniqueclasses):
  sequence_input = Input(shape=(len(uniquewords),))
  dense1 = Dense(128, activation='relu')(sequence_input)
  preds = Dense(len(uniqueclasses), activation='softmax')(dense1)

  model = Model(sequence_input, preds)
  model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['acc'])
  model.summary()
  return model
```

```
model=define_model1(uniqueclasses)
model.fit(array(trainX),array(trainY),batch_size=1000,epochs=10,verbose=1)
```

```
loss, acc = model.evaluate(array(trainX), array(trainY), verbose=0)

print('Train Accuracy: %f' % (acc*100))
```
